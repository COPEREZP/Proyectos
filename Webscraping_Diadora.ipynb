```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

# Función de Scraping de Productos en varias páginas
def scrape_products_multiple_pages():
    # Configuración del ChromeDriver
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Si no quieres ver el navegador en acción
    chrome_options.add_argument("--disable-gpu")  # Deshabilitar GPU (puede ser útil en máquinas sin GPU)
    chrome_options.add_argument("--no-sandbox")  # Evitar errores en entornos Linux (si es necesario)
    
    # Crear el servicio con el webdriver_manager
    service = Service(ChromeDriverManager().install())
    
    # Iniciar el navegador
    driver = webdriver.Chrome(service=service, options=chrome_options)
    
    # Base URL para las páginas
    base_url = "https://www.diadora.com/es/cl/ni%C3%B1os-zapatillas/?srule=sorting-in-season&start={}"
    
    # Inicializar el número de página
    page_number = 0
    found_products = True  # Variable para controlar si hay más productos
    
    # Lista para almacenar los datos de los productos
    product_data = []
    
    while found_products:
        # Construir la URL de la página actual
        url = base_url.format(page_number * 24)
        driver.get(url)
        
        # Esperar que la página cargue completamente
        time.sleep(5)
        
        # Encontrar todos los productos en la página
        productos = driver.find_elements(By.CSS_SELECTOR, "div.card-tile")
        
        # Si no hay productos en la página, terminar el ciclo
        if not productos:
            found_products = False
            print("No se encontraron más productos.")
            break
        
        # Iterar sobre cada producto encontrado y extraer los detalles
        for producto in productos:
            try:
                # Extraer el nombre del producto
                nombre_producto = producto.find_element(By.CSS_SELECTOR, "span.product_title_name").text
                
                # Extraer el precio original
                precio_producto = producto.find_element(By.CSS_SELECTOR, "span.price__basic").text.strip()  # Limpiar espacio extra
                
                # Extraer la URL de la imagen del producto
                url_imagen = producto.find_element(By.CSS_SELECTOR, "img.img-fluid").get_attribute("data-src")
                
                # Extraer la URL del producto
                url_producto = producto.find_element(By.CSS_SELECTOR, "a.itemProduct__link").get_attribute("href")
                
                # Almacenar los datos en la lista
                product_data.append({
                    "Nombre del producto": nombre_producto,
                    "Precio original": precio_producto,
                    "URL de la imagen": url_imagen,
                    "URL del producto": url_producto
                })
                print(f"Producto '{nombre_producto}' extraído exitosamente.")
                
            except Exception as e:
                print("Error al extraer los datos del producto:", e)
        
        # Aumentar el número de página para acceder a la siguiente página
        page_number += 1
    
    # Convertir los datos a un DataFrame de pandas
    df = pd.DataFrame(product_data)
    
    # Exportar los datos a un archivo Excel
    output_path = r"C:\Users\constanza.perez\OneDrive - Colgram\Escritorio\MICHAEL2.xlsx"
    df.to_excel(output_path, index=False)
    print(f"Los datos se han exportado exitosamente a {output_path}")
    
    # Cerrar el navegador
    driver.quit()

# Llamar a la función para realizar el scraping en múltiples páginas y exportar los datos a Excel
scrape_products_multiple_pages()
```
