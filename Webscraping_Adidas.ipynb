```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import random
import os

# Ruta del archivo Excel donde se guardarán los resultados
excel_path = r"C:\Users\constanza.perez\OneDrive - Colgram\Escritorio\DIC.xlsx"

# Lista de IPs SOCKS5 disponibles
ips = [
    "socks5://31.22.7.188:1080",
    "socks5://128.199.59.76:1080",
    "socks5://167.99.50.159:1080",
    "socks5://159.69.117.155:1080",
    "socks5://198.27.82.161:1080",
    "socks5://37.187.73.136:1080",
    "socks5://205.185.119.188:1080",
    "socks5://217.12.201.56:1080",
    "socks5://205.196.220.122:1080",
    "socks5://138.197.69.132:1080",
    "socks5://167.172.23.19:1080",
    "socks5://149.202.23.207:1080",
    "socks5://172.67.187.242:1080",
    "socks5://206.189.15.100:1080",
    "socks5://159.65.116.119:1080",
    "socks5://143.198.237.236:1080",
    "socks5://54.38.242.224:1080",
    "socks5://138.201.140.125:1080"
]

# Función para configurar el navegador con una IP SOCKS5 específica
def get_driver_with_ip(ip):
    chrome_options = Options()
    chrome_options.add_argument("--disable-notifications")
    chrome_options.add_argument(f"--proxy-server={ip}")
    chrome_options.add_argument("--proxy-bypass-list=*")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    return driver

# Función para realizar el scraping y extraer los datos de los productos
def scrape_products():
    # Iniciar la lista de datos de productos
    product_data = []
    
    # Inicializar el contador de página
    page_number = 1

    while True:
        # Seleccionar una IP SOCKS5 aleatoria
        ip = random.choice(ips)
        
        # Iniciar el navegador con la IP SOCKS5 seleccionada
        driver = get_driver_with_ip(ip)
        
        # Generar la URL de la página
        url = f"https://www.adidas.cl/calzado-ninos?start={24 * (page_number - 1)}"
        
        # Acceder a la página
        driver.get(url)
        
        # Esperar un momento para que la página cargue completamente
        time.sleep(3)

        try:
            # Esperar hasta que los productos estén visibles
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "//*[@id='__next']/div[2]/main/section[3]/article[1]"))
            )
            
            # Encontrar todos los productos en la página
            products = driver.find_elements(By.XPATH, "//*[@id='__next']/div[2]/main/section[3]/article")
            
            if len(products) == 0:
                # Si no se encuentran productos, terminar el scraping
                print("No se encontraron más productos. Terminando el scraping.")
                break

            # Iterar a través de todos los productos
            for product in products:
                try:
                    # Extraer los detalles del producto
                    product_name = product.find_element(By.CSS_SELECTOR, '.product-card-description_name__xHvJ2').text
                    product_image = product.find_element(By.CSS_SELECTOR, '.product-card-assets_product-image__OvGsb').get_attribute("src")
                    product_link = product.find_element(By.CSS_SELECTOR, 'a').get_attribute("href")
                    
                    try:
                        discounted_price = product.find_element(By.CSS_SELECTOR, '.gl-price-item[data-testid="primary-price"]').text
                    except:
                        discounted_price = ""

                    try:
                        # Obtener el precio original con el nuevo XPath
                        original_price = product.find_element(By.CSS_SELECTOR, '.gl-price-item[data-testid="true"]').text
                    except:
                        original_price = ""  # Si no tiene precio original, dejamos vacío
                    
                    # Almacenar la información del producto
                    product_data.append({
                        "Nombre": product_name,
                        "Imagen": product_image,
                        "Enlace": product_link,
                        "Precio con Descuento": discounted_price,
                        "Precio Original": original_price
                    })
                except Exception as e:
                    print(f"Error al extraer datos de un producto: {e}")
            
            # Cerrar el navegador
            driver.quit()
            
            # Aumentar el número de página
            page_number += 1
            
        except Exception as e:
            print(f"Error al realizar el scraping: {e}")
            driver.quit()
            break

    # Crear un DataFrame con los resultados
    df = pd.DataFrame(product_data)
    
    # Verificar si el archivo Excel existe
    if os.path.exists(excel_path):
        # Si el archivo Excel existe, leerlo
        existing_df = pd.read_excel(excel_path)
        # Concatenar los datos existentes con los nuevos datos
        combined_df = pd.concat([existing_df, df], ignore_index=True)
        # Escribir los datos combinados en el archivo Excel
        combined_df.to_excel(excel_path, index=False, sheet_name="Productos")
    else:
        # Si el archivo Excel no existe, crear el archivo
        df.to_excel(excel_path, index=False, sheet_name="Productos")
    
    print(f"Se han guardado {len(product_data)} productos en el archivo Excel.")

# Llamar a la función de scraping
scrape_products()
```
