```python
import time
import random
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Lista de proxies SOCKS5 disponibles
ips = [
    "socks5://103.35.108.206:1080",
    "socks5://199.188.93.214:1080",
    "socks5://152.32.186.145:1080",
    "socks5://119.8.111.196:1080",
    "socks5://213.5.17.84:1080"
]

# Ruta del archivo Excel donde se guardarán los resultados
excel_path = r"C:\Users\constanza.perez\OneDrive - Colgram\Escritorio\NIKI.xlsx"

# Función para configurar el navegador con una IP SOCKS5 específica
def get_driver_with_ip(ip):
    chrome_options = Options()
    chrome_options.add_argument("--disable-notifications")
    chrome_options.add_argument(f"--proxy-server={ip}")
    chrome_options.add_argument("--proxy-bypass-list=*")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    return driver

# Función para cerrar los pop-ups que aparecen en la página
def close_popups(driver):
    try:
        # Esperar a que el pop-up aparezca y cerrarlo
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, "modal-close"))).click()
    except Exception as e:
        print("No pop-up found or could not close pop-up:", e)

# Función principal para realizar el scraping de la página
def scrape_products():
    # Seleccionar un proxy aleatorio
    ip = random.choice(ips)
    driver = get_driver_with_ip(ip)
    
    try:
        # Abrir la página de productos de Caffarena
        driver.get("https://www.caffarena.cl/ESCOLAR")
        
        # Esperar a que la página cargue completamente
        time.sleep(5)
        
        # Cerrar posibles pop-ups
        close_popups(driver)
        
        # Buscar los productos dentro de los divs especificados
        product_elements = driver.find_elements(By.CSS_SELECTOR, "div.vtex-search-result-3-x-galleryItem")
        
        product_names = []
        product_links = []
        
        # Iterar sobre cada producto y extraer su nombre y enlace
        for product in product_elements:
            try:
                # Extraer el nombre del producto desde el atributo alt de la imagen
                img_element = product.find_element(By.TAG_NAME, "img")
                product_name = img_element.get_attribute("alt")
                product_names.append(product_name)
                
                # Extraer el enlace del producto
                link_element = product.find_element(By.TAG_NAME, "a")
                product_link = link_element.get_attribute("href")
                product_links.append(product_link)
            except Exception as e:
                print("Error extracting product information:", e)
        
        # Imprimir los nombres de los productos y sus enlaces
        print("Nombres de productos y enlaces encontrados:")
        for name, link in zip(product_names, product_links):
            print(f"Nombre: {name}, Enlace: {link}")
        
        # Ahora vamos a navegar a la página de cada producto
        for link in product_links:
            driver.get(link)
            time.sleep(3)  # Esperar para asegurar que la página cargue completamente
            
            # Aquí podrías extraer más información de la página del producto si lo deseas
            # Ejemplo: Extraer el precio
            try:
                price_element = driver.find_element(By.CSS_SELECTOR, "span.price")
                product_price = price_element.text
                print(f"Precio del producto {link}: {product_price}")
            except Exception as e:
                print(f"No se pudo extraer el precio para {link}")
        
    finally:
        # Cerrar el navegador después de terminar el scraping
        driver.quit()

# Ejecutar la función para realizar el scraping
scrape_products()
```
