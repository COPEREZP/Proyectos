```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def scrape_products():
    # Configura el path del ChromeDriver (asegúrate de tenerlo correctamente instalado)
    service = Service(r'C:\Users\constanza.perez\OneDrive - Colgram\Escritorio\SELENIUM\chromedriver.exe')
    
    # No usamos --headless para poder ver la ventana del navegador
    driver = webdriver.Chrome(service=service)

    # Abre la URL de la página
    url = "https://www.bata.com/cl/infantil/escolar/?start=0&sz=48"
    driver.get(url)

    # Asegura que la página haya cargado completamente
    time.sleep(5)  # Aumento de tiempo para cargar completamente la página (más tiempo)

    # Inicializa una lista para almacenar los datos de los productos
    products_data = []

    try:
        # Espera explícita para asegurarse de que el pop-up de cookies esté presente
        wait = WebDriverWait(driver, 30)
        
        # Intentamos encontrar y cerrar el pop-up de cookies si aparece
        try:
            cookie_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[aria-label='Aceptar']")))
            cookie_button.click()  # Hacemos clic en el botón para cerrar el pop-up
            print("Pop-up de cookies cerrado.")
        except Exception as e:
            print("No se encontró el pop-up de cookies o ya estaba cerrado.")

        # Espera explícita para asegurarse de que los productos estén cargados
        productos = wait.until(lambda driver: driver.execute_script("""
            return document.querySelectorAll("div.cc-col-tile.tile_701M_2023_2016044_1.cc-column-tile");
        """))
        
        # Verificamos si se encontraron productos
        if not productos:
            raise Exception("No se encontraron productos con el selector proporcionado.")
        
        print(f"Se encontraron {len(productos)} productos.")

        # Recorremos todos los productos encontrados
        index = 1  # Empezamos a contar desde el primer producto
        while True:
            try:
                # Si llegamos al producto 24, detenemos el scraping
                if index > 96:
                    break

                # Espera para asegurarse de que el producto esté visible
                wait.until(EC.visibility_of_element_located((By.XPATH, f"//*[@id='product-search-results']/div/div[7]/div[4]/div[{index}]")))
                
                # Extraemos el producto
                producto = driver.find_element(By.XPATH, f"//*[@id='product-search-results']/div/div[7]/div[4]/div[{index}]")
                
                print(f"Extrayendo producto {index}...")

                # Extraemos el nombre del producto
                try:
                    product_name = producto.find_element(By.XPATH, ".//div[2]/div[1]/a/span[2]").text
                    print(f"Nombre del Producto {index}: {product_name}")
                except Exception as e:
                    print(f"Error al extraer el nombre del producto en el producto {index}: {e}")
                    product_name = "No disponible"

                # Extraemos el precio original
                try:
                    original_price = producto.find_element(By.XPATH, ".//div[2]/div[2]/span[3]").text
                    print(f"Precio original del Producto {index}: {original_price}")
                except Exception as e:
                    print(f"Error al extraer el precio original del producto {index}: {e}")
                    original_price = "No disponible"

                # Extraemos el precio final
                try:
                    final_price = producto.find_element(By.XPATH, ".//span[@class='cc-price']").text
                    print(f"Precio final del Producto {index}: {final_price}")
                except Exception as e:
                    print(f"Error al extraer el precio final del producto {index}: {e}")
                    final_price = "No disponible"

                # Extraemos la URL de la imagen
                try:
                    image_url = producto.find_element(By.XPATH, ".//div[1]/a/div/div[1]/div/img").get_attribute("src")
                    print(f"URL de la imagen del Producto {index}: {image_url}")
                except Exception as e:
                    print(f"Error al extraer la URL de la imagen del producto {index}: {e}")
                    image_url = "No disponible"

                # Extraemos la URL de la página del producto
                try:
                    product_url = producto.find_element(By.XPATH, ".//div[1]/a").get_attribute("href")
                    print(f"URL de la página del Producto {index}: {product_url}")
                except Exception as e:
                    print(f"Error al extraer la URL de la página del producto {index}: {e}")
                    product_url = "No disponible"

                # Añadir los datos del producto a la lista
                products_data.append({
                    "Nombre del Producto": product_name,
                    "Precio Original": original_price,
                    "Precio Final": final_price,
                    "URL de la Imagen": image_url,
                    "URL del Producto": product_url
                })

                # Incrementar el índice para el siguiente producto
                index += 1

            except Exception as e:
                print(f"No se pudo encontrar más productos o ha ocurrido un error al acceder al producto {index}. Deteniendo el proceso.")
                break  # Sale del bucle cuando no encuentra más productos

            # Espera para asegurarse de que la página cargue adecuadamente entre productos
            time.sleep(2)

    except Exception as e:
        print(f"Error: {e}")

    # Cierra el navegador después de un tiempo
    time.sleep(10)  # Espera antes de cerrar el navegador para que veas el resultado
    driver.quit()

    # Guardar los productos en un archivo Excel
    try:
        # Crear un DataFrame con los datos
        df = pd.DataFrame(products_data)

        # Cambiar la ruta de salida a 'AYUDA.xlsx'
        output_path = r"C:\Users\constanza.perez\OneDrive - Colgram\Escritorio\TU2.xlsx"
        df.to_excel(output_path, index=False)
        print(f"Los datos de los productos han sido guardados en {output_path}")
    except Exception as e:
        print(f"Error al guardar el archivo Excel: {e}")

# Llamar a la función para realizar el scraping
scrape_products()
```
